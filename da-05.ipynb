```python
# Write a program to print numbers from 1 to 10, but stop if the number is 5.

for i in range(1,11):
    if i==5:
        break
    print(i)
```

```python
l=[1,4,9,"stop",89]

for i in l:
    if i=="stop":
        break
    print(i)
```

```python
# Write a program to skip printing even numbers from 1 to 10.

for i in range(1,11):
    if i%2==0:
        continue
    print(i)
```

```python
# Write a program to print numbers from 0 to 9 using range().

for i in range(10):
    print(i)
```

```python
for i in range(1,6):
    for j in range(1,11):
        num=i*j
        print(num)
    break
```

```python
# Write a program to skip printing even numbers using a while loop.

i=0
while i<10:
    i=i+1
    if i%2==0:
        continue
  
    print(i)
```
