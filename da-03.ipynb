```python
"""
Que6. Differentiate between expression and statement in Python with examples.

An expression is a combination of values, variables, operators, and function calls that Python evaluates to produce a result (usually a value). Expressions return a value and 
can be used wherever Python expects a value.

Key Points about Expressions:
An expression can be a simple value like a number or a variable.
It always evaluates to a result.
It can involve operations, function calls, or logical evaluations.

A statement is a line of code that performs some action. Unlike an expression, a statement doesn't return a value; instead, it carries out an operation or produces side effects, such as changing a variable's value, printing something to the screen, or controlling the flow of the program (e.g., if, while, for, etc.).

Key Points about Statements:
A statement represents an instruction that Python will execute.
It does not return a value like an expression.
Statements typically control the flow of the program or perform actions.
"""
```

```python
#Example:

x = 10  # Statement (assignment)
y = x + 5  # Expression (x + 5 evaluates to 15, and assigns it to y)
if y > 10:  # Statement (conditional)
    print("y is greater than 10")  # Statement (action)
```
