```python
# Print numbers from 1 to 5 using a while loop.

i=1
while(i<6):
    print(i)
    i=i+1
```

```python
# Calculate the sum of numbers from 1 to 10 using a while loop.
sum=0
num=1
while(num<=10):
    sum=sum+num
    num=num+1
print(sum)
```

```python
# Calculate the factorial of a number using a for loop.

num=int(input("Please provide the number"))
fact=1

for i in range(1,num+1):
    fact=fact*i

print("the factorial of given number is ",fact)
    
```

```python
#  Count the number of vowels in a string using a for loop.

s=input("please provide the string")
count=0
for i in s:
    if i=='a' or i=='e' or i=='i' or i=='o' or i=='u':
        count=count+1

print("The number of vowels in the string is ",count)
        
```

```python
# Print a pattern using nested loops.

for i in range(4):
    for j in range (i+1):
        print("*",end="")
    print()

```

```python
# Generate a multiplication table using nested loops.

num=int(input("Give the number for the table"))
i=1
while i <=10:
    table=num*i
    i=i+1
    print(table)



```
