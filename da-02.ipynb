```python
#Calculate the sum, difference, product, and quotient of two numbers.

a=10
b=3

c=a+b   # finding sum
print(c)

c=a-b   # difference
print(c)

c=a*b   # Product/Multiplication
print(c)
 
c=a/b   # Division/finding quotient
print(c)
```

```python
# Perform various assignment operations on a variable.
# Various assignement operations on variable a.

a=3
print(a)

a+=3
print(a)

a-=2
print(a)

a*=5
print(a)

a/=4
print(a)
```

```python
# Compare two numbers and print the results.

a=10
b=20

if a>b:
    print("a is greater than b")
else:
    print("b is greater than a")
```

```python
# Check conditions using logical operators.

x=True 
y=True

if x==True and y==True:
    print("Both and y are true")
elif x==True or y==True:
    print("For x and y, any one of them is true")

print(not x)
```

```python
# Check the identity of variables.

a=20
print(type(a))

b=35.80
print(type(b))

c=True
print(type(c))

s="string"
print(type(s))

d=2+5j
print(type(d))
```

```python
# Perform bitwise operations on any two integers.

x=10
print(x<<2)

y=50
print(y>>3)
```

```python
# Use unary operators to change the sign of a number.

num=5
negative_num=-num
print(negative_num)
```

```python
# Use the ternary operator to assign values based on conditions.

age=18

if age>=18:
    status="adult"
else:
    status="minor"

print(status)
```
