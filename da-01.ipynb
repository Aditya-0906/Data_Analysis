```python
# Convert an integer to a floating-point number.

a=40
b=float(a)
type(b)
```

```python
#  Convert a float to an integer.

b=85.50
a=int(b)
type(a)
```

```python
# Convert an integer to a string.

x=20
y=str(x)
type(y)
```

```python
# Convert a list to a tuple.

l=[1,2,3,4,5]
t=tuple(l)
type(t)
print(t)
```

```python
# Convert a tuple to a list.

t=(10,20,30,40)
l=list(t)
type(l)
```

```python
# Convert a decimal number to binary.

num=20
b=bin(num)
b
```
